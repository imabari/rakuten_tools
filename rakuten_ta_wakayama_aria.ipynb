# 設定

```
# 1TAの距離
N = 150
```

```
# 緯度,経度,TA

data = """\
34.065032,135.258720,28
34.0852011,135.2976488,0
"""
```

# プログラム

```
import io
import pandas as pd
```

```
import folium
```

```
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vRmhTv7hMnKlkpXDCNurEsMkViecXWl0kf4bW43P_6l8TAV3znZtEGbk8MHnOTs57OGjg3Ksv2OMIYx/pub?output=csv"

df0 = (
    pd.read_csv(url, usecols=[1, 2, 3, 6, 7, 10, 12])
    .dropna(how="all")
    .fillna("")
)
```

```
df0
```

```
df0["Placemark Name"] = df0["Placemark Name"].str.strip()
```

```
df1 = df0[~df0["形態"].isin(["屋内局", "Casa"])].copy()
```

```
df1["color"] = "green"
```

```
df1["color"] = df1["color"].mask(df1["eNB-LCID(a,b,c)"].str.contains("撤去"), "gray")
```

```
df1["color"] = df1["color"].mask(df1["eNB-LCID(a,b,c)"].str.contains("未開局"), "red")
```

```
df1
```

```
df2 = pd.read_csv(io.StringIO(data), header=None, names=["lat", "lng", "ta"])
```

```
map = folium.Map(
    tiles=None,
    location=[df2["lat"].mean(), df2["lng"].mean()],
    zoom_start=12,
)
```

```
folium.raster_layers.TileLayer(
    tiles="https://cyberjapandata.gsi.go.jp/xyz/pale/{z}/{x}/{y}.png",
    name='国土地理院',
    attr='&copy; <a href="https://maps.gsi.go.jp/development/ichiran.html">国土地理院</a>'
).add_to(map)
```

```
# Googleマップ標準
folium.raster_layers.TileLayer(
    "https://{s}.google.com/vt/lyrs=m&x={x}&y={y}&z={z}",
    subdomains=["mt0", "mt1", "mt2", "mt3"],
    name="Google Map",
    attr="<a href='https://developers.google.com/maps/documentation' target='_blank'>Google Map</a>",
).add_to(map)
```

```
folium.raster_layers.TileLayer(
    name="楽天モバイル",
    tiles='https://area-map.mobile.rakuten.co.jp/dsd/geoserver/4g4m/mno_coverage_map/gwc/service/gmaps?LAYERS=mno_coverage_map:all_map&FORMAT=image/png&TRANSPARENT=TRUE&x={x}&y={y}&zoom={z}&update=20220404',
    fmt='image/png',
    attr="楽天モバイルエリア",
    tms=False,
    overlay=True,
    control=True,
    opacity=1,
).add_to(map)
```

```
folium.raster_layers.TileLayer(
    name="楽天モバイル5G",
    tiles="https://area-map.mobile.rakuten.co.jp/dsd/geoserver/5g/mno_coverage_map/gwc/service/gmaps?LAYERS=mno_coverage_map:all_map&FORMAT=image/png&TRANSPARENT=TRUE&x={x}&y={y}&zoom={z}&update=20220404",
    fmt="image/png",
    attr="楽天モバイルエリア",
    tms=False,
    overlay=True,
    control=True,
    opacity=1,
    show=False,
).add_to(map)
```

# map

```
fg0 = folium.FeatureGroup(name="基地局").add_to(map)
fg1 = folium.FeatureGroup(name="サークル").add_to(map)
fg2 = folium.FeatureGroup(name="観測地点").add_to(map)
fg3 = folium.FeatureGroup(name="TA").add_to(map)
```

```
for i, r in df1.iterrows():

    fg0.add_child(
        folium.Marker(
            location=[r["latutude"], r["longitude"]],
            popup=folium.Popup(f'<p><a href="https://www.google.com/maps?layer=c&cbll={r["latutude"]},{r["longitude"]}" target="_blank">{r["Placemark Name"]}</a></p><p>{r["eNB-LCID(a,b,c)"]}</p>', max_width=300),
            tooltip=f'{r["Placemark Name"]}',
            icon=folium.Icon(color=r.color, icon="signal"),
        )
    )

    if r.color != "gray":

        fg1.add_child(
            folium.Circle(
                location=[r["latutude"], r["longitude"]],
                radius=780,
                color="black",
                weight = 1,
                fill=True,
                fill_opacity=0.5,
            )
        )
```

```
for i, r in df2.iterrows():

    fg2.add_child(
        folium.Marker(
            location=[r.lat, r.lng],
            icon=folium.Icon(color="blue"),
        )
    )

    if r.ta > 0:

        fg3.add_child(
                folium.Circle(
                location=[r.lat, r.lng],
                radius=r.ta * N,
                color="yellow",
            )
        )
```

```
lat, lng = 33.967677, 133.280802

folium.Marker(
    location=[lat, lng],
).add_to(map)
```

```
folium.LayerControl().add_to(map)
```

```
from folium import plugins
```

```
# DRAW
folium.plugins.Draw(draw_options = {'polygon': False,'rectangle': False,'circlemarker': False}).add_to(map)
```

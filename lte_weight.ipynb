```python
import io
import pandas as pd
```

```python
data = """\
name,lat,lng,rsrp
白川公園,35.1630791210542,136.89975760306444,-120
名古屋城,35.18558653662987,136.89978515031999,-128
金山駅南口の前の通り,35.14212085043628,136.90051553138235,-67
金山南ビル美術館棟,35.142593722830306,136.9001809856621,-64
金山のマック,35.14342670545131,136.90276490298538,-88
イオン金山店,35.144921269970396,136.90080787489498,-87
アスナル金山の真ん中,35.14395421757375,136.90083770262473,-99
古沢公園,35.146594509673,136.90180932179103,-91
東別院駅,35.14987388545931,136.90462585198628,-107
"""
df = pd.read_csv(io.StringIO(data))
```

```python
df["weight"] = df["rsrp"].apply(lambda x: ((1.0 / (x + 5.0) ** 2) * 10000) ** 2)
```

```python
df.to_csv(index=False)
```

```python
# Map -140: 0.3, -105: 1.0, -89: 2.0, -55: 16.0, -43: 48.0
```

```python
# 重み付き平均により推定位置を計算
weighted_lat_sum = (df["lat"] * df["weight"]).sum()
weighted_lng_sum = (df["lng"] * df["weight"]).sum()
total_weight = df["weight"].sum()
```

```python
estimated_lat = weighted_lat_sum / total_weight
estimated_lng = weighted_lng_sum / total_weight
```

```python
print(f"推定位置の緯度: {estimated_lat}")
print(f"推定位置の経度: {estimated_lng}")
```

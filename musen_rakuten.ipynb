# 選択

```python
#@title 都道府県名 { display-mode: "form" }
pref = "\u5317\u6D77\u9053" #@param ["北海道","青森県","岩手県","宮城県","秋田県","山形県","福島県","茨城県","栃木県","群馬県","埼玉県","千葉県","東京都","神奈川県","山梨県","新潟県","長野県","富山県","石川県","福井県","岐阜県","静岡県","愛知県","三重県","滋賀県","京都府","大阪府","兵庫県","奈良県","和歌山県","鳥取県","島根県","岡山県","広島県","山口県","徳島県","香川県","愛媛県","高知県","福岡県","佐賀県","長崎県","熊本県","大分県","宮崎県","鹿児島県","沖縄県"]
```

```python
#@title 種類 { vertical-output: true }
kind = "macro" #@param ["macro", "femto"]
```

```python
pref
```

```python
kind
```

```python
area = pref_area.get(pref)
```

```python
# 総合通信局
p_csv = pathlib.Path("rakuten-main", "csv", kind, area)
```

```python
dfs = []
```

```python
# CSVファイルをdataframeに変換
for p in p_csv.glob("*.csv"):

    tmp = pd.read_csv(p, index_col=[0, 1]).T
    dfs.append(tmp)

# 結合
df0 = pd.concat(dfs).fillna(0).astype(int)
```

```python
# indexをdatetimeに変換
df0.index = pd.to_datetime(df0.index)
```

```python
# ソート
df0.sort_index(inplace=True)
```

# 日別開局数計

```python
# 指定都道府県抽出
df1 = df0[[pref]].droplevel("pref", axis=1)
```

```python
df1
```

```python
df1.to_csv("data.csv", encoding="utf_8_sig")
```

# 日別新規開局数

```python
df2 = df1.diff().dropna(how="all").fillna(0).astype(int)
```

```python
df2.to_csv("diff.csv", encoding="utf_8_sig")
```
